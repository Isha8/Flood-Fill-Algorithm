```python
#function fill_neighbors works in a similar way as the flood fill algortihm. It takes 4 input parameters:
#%%!canvas is the world (list of list) with as many different regions as required. 
#%%!r and c are the row and column index of the point from where the filling is to be started
#%%!new_value(can be int, decimal, char, or string )is the value with which the neigborhood is to be replaced
#It returns a new world with the replaced neigborhood.

def fill_neighbors(canvas,r,c,new_value):

    list_of_all= []  #final list of all pixels (may have redundant)
    current = [[r,c]] #list of neighbors of current pixel initialised with [r,c]
    done = [] #list of traversed pixels
    a = canvas[r][c] 

    while current:
        curr = current.pop()
        r,c = curr[0],curr[1]
        list_of_all.append(curr)
        done.append(curr)


        if (r < len(canvas) - 1 ):
            if (canvas[r+1][c] == a):
                if ([r+1,c] not in current and [r+1,c] not in done):
                    current.append([r+1,c])
                list_of_all.append([r+1,c])

            if (r >= 1):
                if (canvas[r-1][c] == a):
                    if ([r-1,c] not in current and [r-1,c] not in done):
                        current.append([r-1,c])
                    list_of_all.append([r-1,c])
                    
        elif(r >=1):
            if (canvas[r-1][c] == a):
                    if ([r-1,c] not in current and [r-1,c] not in done):
                        current.append([r-1,c])
                    list_of_all.append([r-1,c])
            
            if (r < len(canvas) - 1 ):
                if (canvas[r+1][c] == a):
                    if ([r+1,c] not in current and [r+1,c] not in done):
                        current.append([r+1,c])
                    list_of_all.append([r+1,c])




        if (c < len(canvas[0])-1):
            if (canvas[r][c+1] == a):
                if ([r,c+1] not in current and [r,c+1] not in done):
                    current.append([r,c+1])
                list_of_all.append([r,c+1])

            if (c >= 1):
                if (canvas[r][c-1] == a):
                    if ([r,c-1] not in current and [r,c-1] not in done):
                        current.append([r,c-1])
                    list_of_all.append([r,c-1])
                    
        elif(c>=1):
            if (canvas[r][c-1] == a):
                    if ([r,c-1] not in current and [r,c-1] not in done):
                        current.append([r,c-1])
                    list_of_all.append([r,c-1])

            if (c < len(canvas[0])-1):
                if (canvas[r][c+1] == a):
                    if ([r,c+1] not in current and [r,c+1] not in done):
                        current.append([r,c+1])
                    list_of_all.append([r,c+1])

    canvas_out = canvas 
    for each in list_of_all:
        canvas_out[each[0]][each[1]] = new_value
    return (canvas_out)

    

if __name__ == "__main__":    
    canvas = [[0,1,1,2,0,0],
              [1,1,2,2,1,1],
              [0,0,0,1,1,2],
              [1,0,1,2,1,2],
              [2,2,2,1,1,1],
              [1,1,1,1,0,0]]
    r=2
    c=3
    new_value=8
    after_filling = fill_neighbors(canvas,r,c,new_value)
    for each in after_filling:
        print(each)
```
